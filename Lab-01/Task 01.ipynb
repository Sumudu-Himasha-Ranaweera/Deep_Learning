## 02.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the parameters for the exponential distribution
scale_parameter = 1.0

# Generate a random 100000x1 array from an exponential distribution
exp_random_array = np.random.exponential(scale=scale_parameter, size=(100000, 1))

# Generate a random 100000x1 array from a uniform distribution
uniform_random_array = np.random.rand(100000, 1)

# Generate a random 100000x1 array from a normal distribution
normal_random_array = np.random.normal(size=(100000, 1))

# Create a histogram for the exponential distribution
plt.hist(exp_random_array, density=True, bins=50, histtype="step", color="blue", label="Exponential")

# Overlay histograms for uniform and normal distributions
plt.hist(uniform_random_array, density=True, bins=50, histtype="step", color="green", label="Uniform")
plt.hist(normal_random_array, density=True, bins=50, histtype="step", color="red", label="Normal")

plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

## 03. 

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate the corresponding z values using the function Z = X**2 + Y**2
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set the title of the plot
ax.set_title('Z = X**2 + Y**2')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read dataset
df = pd.read_csv('C:/Users/himashara/Documents/SLIIT/2 SEMESTER/Deep Learning - SE4050 [2022JUL]/LABS/Lab 01/Pokemon.csv', index_col=0, encoding='latin1')
```

```python
# Select the specific columns of interest
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
correlation_pearson = df[features].corr(method='pearson')
correlation_spearman = df[features].corr(method='spearman')
```

```python
# Set up the figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot Pearson correlation heatmap
sns.heatmap(correlation_pearson, annot=True, cmap='coolwarm', center=0, ax=axes[0])
axes[0].set_title('Pearson Correlation')

# Plot Spearman correlation heatmap
sns.heatmap(correlation_spearman, annot=True, cmap='coolwarm', center=0, ax=axes[1])
axes[1].set_title('Spearman Correlation')

# Adjust layout
plt.tight_layout()
plt.show()
```
