```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset into a DataFrame
df = pd.read_csv('C:/Users/himashara/Documents/SLIIT/2 SEMESTER/Deep Learning - SE4050 [2022JUL]/LABS/Lab 01/Au_nanoparticle_dataset.csv')
```

```python
print(df.dtypes)
```

```python
# Create a new dataframe with selected columns
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

```python
# Display the first 20 samples
print(new_df.head(20))
```

```python
# Calculate mean, standard deviation, and quartile values
stats = new_df.describe()
print(stats)
```

```python
# Plot histograms of each feature in a 1x4 layout
plt.figure(figsize=(16, 4))
for i, col in enumerate(new_df.columns):
    plt.subplot(1, 4, i + 1)
    sns.histplot(new_df[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# Plot histograms of each feature in a 1x4 layout
plt.figure(figsize=(16, 4))
for i, col in enumerate(new_df.columns):
    plt.subplot(1, 4, i + 1)
    sns.histplot(new_df[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# Visualize scatter plots and histograms using pairplot
sns.pairplot(new_df)
plt.show()
```

```python
# Enhance the pairplot
g = sns.PairGrid(new_df)
g.map_upper(sns.histplot)  # Bivariate histogram
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
plt.show()
```
